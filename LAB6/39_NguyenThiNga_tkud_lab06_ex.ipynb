## <u>Bài 1</u>

Green M&M Candies liên quan đến Dataset 18 trong file excel. Tìm tỉ lệ mẫu của M&M có màu xanh lá cây (green). Sử dụng kết quả để xây dựng 1 ước lượng khoảng tin cậy 95% của tổng thể M&M có màu xanh.

```python
df_M_M = pd.read_excel('18_M_M.xls')
df_M_M
```

```python
# Count the number of green candies
green_count = df_M_M['Green'].count()
print(green_count)

# Calculate the sample proportion
sample_proportion = green_count / len(df_M_M)

# Calculate the standard error
standard_error = np.sqrt(sample_proportion * (1 - sample_proportion) / len(df_M_M))

# Calculate the margin of error
margin_of_error = 1.96 * standard_error

# Calculate the lower and upper bounds of the confidence interval
lower_bound = sample_proportion - margin_of_error
upper_bound = sample_proportion + margin_of_error

# Print the sample proportion and confidence interval
print("Sample Proportion:", sample_proportion)
print("Confidence Interval (95%):", (lower_bound, upper_bound))
```

## <u>Bài 3</u>

Lượng mưa ở Boston: liên quan đến Dataset 14 trong file excel, và quan tâm đến các ngày với các giá trị lượng mưa khác nhau từ 0 đến các ngày có mưa có giá trị lượng mưa lớn hơn 0.

- Xây dựng ước lượng khoảng tin cậy 95% cho tỉ lệ mưa trong các ngày Thứ Tư và 
- Xây dựng ước lượng khoảng tin cậy 95% cho tỉ lệ mưa trong các ngày Chủ Nhật.

```python
df_bostrain= pd.read_excel('14_BOSTRAIN.xls')
df_bostrain
```

```python
# Calculate the proportion of rain on Wednesdays
wednesday_rain_proportion = df_bostrain['WED'].mean()

# Calculate the sample size
sample_size = len(df_bostrain)

# Calculate the standard error
standard_error = np.sqrt(wednesday_rain_proportion * (1 - wednesday_rain_proportion) / sample_size)

# Calculate the margin of error
margin_of_error = 1.96 * standard_error

# Calculate the lower and upper bounds of the confidence interval
lower_bound = wednesday_rain_proportion - margin_of_error
upper_bound = wednesday_rain_proportion + margin_of_error

# Print the confidence interval
print("Confidence Interval (95%):", (lower_bound, upper_bound))
```

```python
sunday_rain_proportion = df_bostrain['SUN'].mean()
sample_size = len(df_bostrain)

standard_error = np.sqrt(sunday_rain_proportion * (1 - sunday_rain_proportion) / sample_size)
margin_of_error = 1.96 * standard_error

lower_bound = sunday_rain_proportion - margin_of_error
upper_bound = sunday_rain_proportion + margin_of_error

(lower_bound, upper_bound)
```

## <u>Bài 4</u>

Tổng số tiền phim: liên quan đến Dataset 9 trong file excel. Xây dựng ước tính khoảng thời gian tin cậy 95% của tổng số tiền trung bình cho quần thể của tất cả các phim. Giả định rằng độ lệch chuẩn của quần thể được biết là 100 triệu đô la. 

```python
df_movies = pd.read_excel('09_MOVIES.xls')
df_movies
```

```python

sample_mean = df_movies['Gross'].mean()
sample_size = len(df_movies)
population_std = 100

z = stats.norm.ppf(0.975)  # 95% confidence level, two-tailed test

margin_of_error = z * (population_std / (sample_size ** 0.5))

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

(lower_bound, upper_bound)
```

## <u>Bài 5</u>

Điểm đánh giá tín dụng FICO: liên quan đến Dataset 24 trong file excel. Xây dựng ước lượng khoảng tin cậy 99% của điểm FICO trung bình cho quần thể. Giả sử độ lệch chuẩn của quần thể là 92,2.

```python
df_fico = pd.read_excel('24_FICO.xls')
df_fico
```

```python
sample_mean = df_fico.mean()
sample_size = len(df_fico)
population_std = 92.2

z = stats.norm.ppf(0.995)  # 99% confidence level, two-tailed test

margin_of_error = z * (population_std / (sample_size ** 0.5))

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

(lower_bound, upper_bound)
```
