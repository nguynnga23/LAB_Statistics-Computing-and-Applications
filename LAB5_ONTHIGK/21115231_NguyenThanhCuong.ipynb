```python
# Tìm hiểu thông tin về dữ liệu:
# a.	(0.5đ) Dữ liệu bao gồm những thuộc tính nào? Liệt kê theo nhóm các thuộc tính định tính, các thuộc tính định lượng và chọn tham số thống kê mô tả phù hợp để đo độ tập trung dữ liệu của mỗi nhóm.

# b.	(1.5đ) Có bao nhiêu loại công ty (company type) trong khảo sát? Liệt kê các loại công ty đó.
CompanyType_counts = df.groupby(['Company type']).size()
# So luong loai cong ty
print( CompanyType_counts.shape[0])
# Liet ke loai cong ty
print(CompanyType_counts)
```

```python
# Cho biết tỷ lệ nam và nữ xuất hiện trong khảo sát? Chọn loại đồ thị phù hợp trực quan để làm nổi bật so sánh của 2 tỷ lệ này
df.head(5)

gender_counts = df['Gender'].value_counts().reset_index(name="counts")
print(gender_counts)

plt.bar(gender_counts['index'], gender_counts['counts'])
plt.show()
```

```python
# Để so sánh mức lương của nam và nữ trong ngành IT:
# a.	Tính đại lượng đặc trưng cho độ tập trung và độ phân tán về mức lương của nam và nữ. Bình luận kết quả tính được.
male_salary = df[df['Gender'] == "M"]
female_salary = df[df['Gender'] == "F"]


print('1.Do tap trung')
print('1.1. Trung binh luong')
print("Cua nam: " +str(male_salary.mean()['Current Salary']))
print("Cua nu: " +str(female_salary.mean()['Current Salary']))
print('1.2. Trung vi luong')
print("Cua nam: " +str(male_salary.median()['Current Salary']))
print("Cua nu: " +str(female_salary.median()['Current Salary']))


print('2.Do phan tan')
print('2.1. Ky vong')
print("Cua nam: " + str(df.var()['Current Salary']))
print("Cua nu:" + str(female_salary.var()['Current Salary']))
print('2.2. Do sai lech')
print("Cua nam: " + str(df.std()['Current Salary']))
print("Cua nu: " + str(female_salary.std()['Current Salary']))

print('Binh luan: muc luong cua nam thuong cao hon muc luong cua nu')

# b.	Hãy chọn đồ thị phù hợp để trực quan làm nổi bật sự so sánh mức lương nam và nữ. Đưa ra bình luận của anh/chị. Kết quả này có tương đồng với bình luận của anh/chị ở trên không?

```
