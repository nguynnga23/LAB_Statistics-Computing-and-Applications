1. Viết chương trình nhập vào số nguyên n (n>0). In ra các số nguyên tố từ 1..n


* Số nguyên tố là số chỉ có 2 ước là 1 và chính nó
* Các bước kiểm tra:
    - Kiểm tra số dương : số nguyên tố phải là một số nguyên dương
    - Kiểm tra trường hợp n >= 1 : True : không là số nguyên tố
    - Duyệt qua các ước từ [2 - căn bậc 2 n]:
        + Nếu n có 1 ước "i" trong khoảng trên thì n không phải là số nguyên tố
        + Ngược lại, n là số nguyên tố

```python
# Hàm kiểm tra 1 số nguyên dương có phải là số nguyên tố không
def isPrime(n):
    if(n <= 1):
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# In ra các số nguyên tố từ 1..n
n = int(input("n = "))
for i in range (1, n+1):
    if isPrime(i):
        print(i)
```

2.Viết chương trình nhập vào số nguyên n có 3 chữ số. Tính tổng các chữ số của n

```python
n = int(input("n (số nguyên có 3 chữ số) = "))

# Kiểm tra xem số nhập vào có đúng là một số nguyên có 3 chữ số hay không
if 100 <= n <= 999:
    donVi = n % 10  #Lấy phần dư của n
    chuc = (n // 10) % 10 
    tram = n // 100
    tong = donVi + chuc + tram

    print(f"Tổng các chữ số của {n} là : {tong}")
else:
    print("Số không hợp lệ. Hãy nhập số nguyên có 3 chữ số.")
```

3. Viết chương trình nhập vào một dãy các số nguyên được cách nhau bởi khoảng trắng (VD: 32 8 4 6 9...) lưu vào python ở dạng mảng. Thực hiện

- Xuất mảng
- Tính tổng các phần tử của mảng
- Tìm số lớn nhất
- Tìm số bé nhất
- Sắp xếp mảng theo thứ tự giảm dần
- In ra các số chẵn trong mảng

```python
#Lưu dãy số nguyên được cách nhau bởi khoảng trắng ở dạng mạng
def print_arr(arr = []):
    print(f"Mảng : {arr}")
    
#Tính tổng các phần tử của mảng
def total_of_arr(arr = []):
    total = 0
    for i in arr:
        total += i
    return total

#Tìm số lớn nhất
def max_of_arr(arr = []):
    max = arr[0]
    for i in arr:
        if i > max:
            max = i
    return max

#Tìm số nhỏ nhất
def min_of_arr(arr = []):
    min = arr[0]
    for i in arr:
        if i < min:
            min = i
    return min

#Sắp xếp mảng theo thứ tự giảm dần
def sort_arr(arr = []):
    arr.sort(reverse = True)
    return arr

#In ra các số chẵn trong mảng
def print_even_number(arr = []):
    arr_evenNum = []
    for i in arr:
        if i % 2 == 0:
            arr_evenNum.append(i)
    return arr_evenNum

# Test
input_str = input("Nhập dãy số nguyên (cách nhau bởi khoảng trắng) : ")
arr_int = [int(num) for num in input_str.split()]

print_arr(arr_int)
print(f"Tổng các phần tử trong mảng là : {total_of_arr(arr_int)}")
print(f"Max : {max_of_arr(arr_int)}")
print(f"Min : {min_of_arr(arr_int)}")
print(f"Mảng giảm dần : {sort_arr(arr_int)}")
print(f"Các số chẵn trong mảng : {print_even_number(arr_int)}")
```

4. Viết chương trình yêu cầu người dùng nhập họ tên. Hiển thị tên người dùng sau khi đã chuẩn hóa. Biết rằng tên đã chuẩn hóa là:
- Ký tự đầu viết hoa
- Không có khoảng trắng thừa
- Không có ký tự đặc biệt

```python
# Hàm chuẩn hóa tên
def standardize_name(name_input):
    # Kí tự đầu viết hoa , không có khoảng trắng thừa
    name = ' '.join(name_input.split()).title()
    # Không có ký tự đặc biệt
    if any(not ch.isalnum() and ch != ' ' for ch in name_input):
        print(("Tên không được chứa ký tự đặc biệt!"))
        return None
    return name

# Test
full_name = input("Input your fullname: ")
standard_name = standardize_name(full_name)
if  standard_name is not None:
    print(f"Your standard name are {standard_name} ")
```

5. Viết chương trình nhập vào một số nguyên dương (n>0). Cho biết n có phải là số đối xứng hay không? Biết số đối xứng là số viết xuôi hay viết ngược đều như nhau

```python
n = input("Nhập một số nguyên dương: ")
if n == n[::-1]:
    print(n, "là số đối xứng")
else:
    print(n, "không phải là số đối xứng")
```
