### 1, (1 điểm) Đọc dữ liệu

Đọc file dữ liệu  và thực hiện các yêu cầu sau:

a. Hiển thị 15 dòng đầu tiên

b. Cho biết dữ liệu có kích thước bao nhiêu (bao nhiêu dòng, bao nhiêu cột)



```python
import pandas as pd
movies = pd.read_csv('movies_metadata.csv', encoding='latin-1')
# Hiển thị 15 dòng đầu tiên
movies.head(15)
# Cho biết dữ liệu có kích thước bao nhiêu (bao nhiều dòng, bao nhiêu cột)
movies.shape
```

### 2, (1 điểm) Kiểm tra dữ liệu

a, Kiểm tra dữ liệu bị thiếu: sử dụng hàm `isnull()` và `sum()`

b. Liệt kê các dòng không có dữ liệu

```python
# Kiểm tra dữ liệu bị thiếu :
movies.isnull().sum()
# Liệt kê các dòng không có dữ liệu
missing_rows = movies[movies.isnull().any(axis=1)]
missing_rows
```

### 3, (1 điểm) Cho biết có bao nhiêu phim sản xuất bằng tiếng Anh


```python
english_movies_count = movies[movies['original_language'] == 'en'].shape[0]
english_movies_count
```

### 4, (2 điểm) Tính giá trị thống kê trung bình, trung vị độ lệch chuẩn hai thuộc tính

* Ngân sách (budget) 

* Doanh thu (revenue)

```python
budget_mean = movies['budget'].mean()
budget_median = movies['budget'].median()
budget_std = movies['budget'].std()

revenue_mean = movies['revenue'].mean()
revenue_median = movies['revenue'].median()
revenue_std = movies['revenue'].std()

print("Thống kê cho thuộc tính 'budget': ")
print("- Trung bình:", budget_mean)
print("- Trung vị:", budget_median)
print("- Độ lệch chuẩn:", budget_std)

print("\nThống kê cho thuộc tính 'revenue': ")
print("- Trung bình:", revenue_mean)
print("- Trung vị:", revenue_median)
print("- Độ lệch chuẩn:", revenue_std)
```

### 5, (2 điểm) Liệt kê top ten

* 10 phim có ngân sách đầu tư (budget) nhiều nhất

* 10 phim có doanh thu cao nhất


```python
# Liệt kê top 10 phim có ngân sách đầu tư nhiều nhất
top_budget_movies = movies.nlargest(10, 'budget')[['title', 'budget']]
print("Top 10 phim có ngân sách đầu tư nhiều nhất:")
print(top_budget_movies)

top_revenue_movies = movies.nlargest(10, 'revenue')[['title', 'revenue']]
print("\nTop 10 phim có doanh thu cao nhất:")
print(top_revenue_movies)
```

### 6, (1 điểm) Liệt kê 10 phim có lợi nhuận cao nhất


```python
# Tính lợi nhuận
movies['profit'] = movies['revenue'] - movies['budget']
top_profit_movies = movies.nlargest(10, "profit")[['title', 'profit']]
print("Top 10 phim có lợi nhuận cao nhất:")
print(top_profit_movies)
```

### 7, (1 điểm) Vẽ đồ thị phân phối tần số về thời lượng của bộ phim (`runtime`)


```python
import matplotlib.pyplot as plt
valid_runtimes = movies['runtime'].dropna()
plt.figure(figsize=(15, 6))
plt.hist(valid_runtimes, bins=30, color="pink", edgecolor = "black")
plt.title("Phấn phôi tần số của thời lượng phim")
plt.xlabel("Thời lượng (phút)")
plt.ylabel("Tần số")
plt.grid(True)
plt.show()
```

### 8, (1 điểm) Vẽ đồ thị boxplot về doanh thu (revenue). Dựa vào đồ thị trên bạn hãy cho nhận xét về doanh thu của phim `Jumanji`


```python
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.boxplot(data = movies, x='revenue')
plt.title("Boxplot về doanh thu của các bộ phim")
plt.xlabel("Doanh thu")
plt.show()
```
