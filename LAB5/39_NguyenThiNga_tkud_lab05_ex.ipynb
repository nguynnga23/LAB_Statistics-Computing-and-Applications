### Bài 1: Tạo quần thể mô phỏng

Bạn hãy tạo một mẫu gồm 10000 phần tử mô phỏng chiều cao của nam thanh niên có giá trị từ 120cm - 200cm. Bạn hãy lưu kết quả vào biến **POP**.

Tính:
a. Chiều cao trung bình (kỳ vọng) của quần thể và độ lệch chuẩn về chiều cao của quần thể.

b. Tính tỷ lệ người cao trong quần thể, biết rằng thanh niên có chiều cao từ 180 trở lên được gọi là cao.

c. Vẽ histogram về chiều cao của quần thể. Theo bạn quần thể có phân phối chuẩn hay không?

d. Thử vẽ histogram và đồ thị hàm mật độ của phân phối chuẩn sử dụng tham số loc và scale bằng với kỳ vọng và độ lệch chuẩn của quần thể.

```python
MIN_HEIGHT = 120
MAX_HEIGHT = 200
POP = np.random.randint(MIN_HEIGHT, MAX_HEIGHT, 10000)
# a. Chiều cao trung bình (kỳ vọng) của quần thể và độ lệch chuẩn về chiều cao của quần thể.
mean_height = np.mean(POP)
std_height = np.std(POP)
print("Chiều cao trung bình (kì vọng) của quần thể là:", mean_height)
print("Độ lệch chuẩn về chiều cao của quẩn thể là:", std_height)
```

```python
# b. Tính tỷ lệ người cao trong quần thể, biết rằng thanh niên có chiều cao từ 180 trở lên được gọi là cao.
count = 0
# for i in range(n):
#     if POP[i] > 180:
#         count += 1
print(f"Tỉ lệ người cao trong quần thể là :{np.sum(POP>180)/len(POP):.4f}")
```

```python
# c. Vẽ histogram về chiều cao của quần thể. Theo bạn quần thể có phân phối chuẩn hay không?
plt.hist(POP,bins=50, density = True, edgecolor="black")
plt.xlabel("Chiều cao")
plt.ylabel("Mật độ")
plt.title("Chiều cao quần thể")
plt.show()

print("Phân tích phân phối : ")
# Tính độ lệch
skewness = np.mean((POP - mean_height) ** 3) / std_height ** 3
# In kết quả
print(f"Độ lệch: {skewness:.2f}")
# Phân tích
if abs(skewness) < 2:
    print("=> Phân phối chiều cao có vẻ đối xứng.")
else:
    print("=> Phân phối chiều cao có vẻ lệch.")
    
```

```python
# Thử vẽ histogram và đồ thị hàm mật độ của phân phối chuẩn sử dụng tham số loc và scale bằng với kỳ vọng và độ lệch chuẩn của quần thể.

x = np.linspace(MIN_HEIGHT, MAX_HEIGHT, 1000)
y = norm.pdf(x, loc=mean_height, scale=std_height)
plt.plot(x, y, color="r", label="Phân phối chuẩn")
plt.title('HÀM MẬT ĐỘ PHÂN PHỐI CHUẨN CỦA TỔNG THỂ')
plt.xlabel("Chiều cao")
plt.ylabel("Mật độ")
plt.legend()
plt.show()
```

=> Mô phỏng này cho thấy chiều cao của nam thanh niên có xu hướng phân phối chuẩn với chiều cao trung bình 165.34 cm và độ lệch chuẩn 14.23 cm. Khoảng 20.47% nam thanh niên được xem là cao (chiều cao từ 180 cm trở lên).

### Bài 4: Khảo sát phân phối của trung bình mẫu

a. Thực hiện bài 2 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

b. Thực hiện bài 3 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

c. Vẽ histogram của hai câu a, b trong cùng một hình để so sánh. Theo bạn hình dạng của histogram thay đổi nói lên điều gì?

Gợi ý: tạo một danh sách rỗng để chứa các trung bình mẫu tính được, sử dụng hàm append() để thêm giá trị trung bình vào danh sách sau mỗi lần tính.

```python
# a. Thực hiện bài 2 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.
sample_20_means = []
for _ in range(100):
    sample = np.random.choice(POP, size=20)
    sample_20_means.append(np.mean(sample))

plt.hist(sample_20_means, bins=20, edgecolor="black")
plt.xlabel("Trung bình mẫu")
plt.ylabel("Tần suất")
plt.show()
```

```python
# b. Thực hiện bài 3 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

sample_100_means = []
for _ in range(100):
    sample = np.random.choice(POP, size=100)
    sample_100_means.append(np.mean(sample))

plt.hist(sample_100_means, bins=20, edgecolor="black")
plt.xlabel("Trung bình mẫu")
plt.ylabel("Tần suất")
plt.show()
```

```python
# c. Vẽ histogram của hai câu a, b trong cùng một hình để so sánh. Theo bạn hình dạng của histogram thay đổi nói lên điều gì?
plt.subplot(121)
plt.hist(sample_20_means, bins=20, label="Mẫu 20", color='pink', edgecolor="black")
plt.xlabel("Trung bình mẫu")
plt.ylabel("Tần suất")

plt.subplot(122)
plt.hist(sample_100_means, bins=20, label="Mẫu 100", edgecolor="black")
plt.xlabel("Trung bình mẫu")
plt.ylabel("Tần suất")

plt.legend()
plt.show()
```

So sánh hình dạng của hai histogram:
- Độ cao: Histogram của dân số nào cao hơn cho thấy dân số đó có nhiều khả năng có trung bình mẫu gần với trung bình của toàn bộ quần thể hơn.
- Độ rộng: Histogram nào có độ rộng (phân tán) hẹp hơn cho thấy dân số đó có độ đồng đều cao hơn về trung bình mẫu.
- Hình dạng: Histogram nào có dạng đối xứng hơn cho thấy dân số đó có khả năng phân phối chuẩn cao hơn.

### Bài 6: Khảo sát phương sai mẫu
